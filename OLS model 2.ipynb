```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('Advertising Budget and Sales (1).csv')
df
```

```python
x=df.drop(labels=['Unnamed: 0','Sales ($)'],axis=1)
y=df['Sales ($)']
X=sm.add_constant(x)
```

```python
model=sm.OLS(y,X).fit()
print(model.summary())
```
