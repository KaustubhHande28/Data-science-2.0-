## Welcome to NumPy!

NumPy (Numerical Python) is an open source Python library that’s used in almost every field of science and engineering. It’s the universal standard for working with numerical data in Python.

The NumPy library contains multidimensional array and matrix data structures.

It provides ndarray, a homogeneous n-dimensional array object, with methods to efficiently operate on it. NumPy can be used to perform a wide variety of mathematical operations on arrays.

### Generating Random Numbers

np.random.randint is a function provided by the NumPy library in Python for generating random integers.


`np.random.randint(low, high=None, size=None, dtype=int)`

*Generates random new output, every tim we run.

* low: (Required) The lowest (inclusive) integer to be drawn from the distribution.
* high: (Optional) The highest (exclusive) integer to be drawn from the * distribution. If omitted, high is set to None, and the range is assumed to be from 0 to low.
* size: (Optional) The shape of the output array. If None, a single integer will be generated.
* dtype: (Optional) The data type of the output array. Default is int.

```python
import numpy as np
```

```python
random_int =np.random.randint(10)

random_int = np.random.randint(5,10)
random_int

random_array = np.random.randint(5,10, size=(3,3))
random_array

random_array = np.random.randint(-10,10, size=(3,3))
random_array
```

####Reverse an Array:

NumPy’s np.flip() function allows you to flip, or reverse, the contents of an array along an axis. When using np.flip(), specify the array you would like to reverse and the axis. If you don’t specify the axis, NumPy will reverse the contents along all of the axes of your input array.


```python
random_array=np.flip(random_array,axis=1)
random_array
```

```python
random_array=np.flip(random_array,axis=0)
random_array
```

#### flattening multidimensional arrays

This section covers .flatten(), ravel()
There are two popular ways to flatten an array: .flatten() and .ravel().

```python
flattened_array = random_array.flatten()
flattened_array
```

```python
flattened_array[3] = 5
print(flattened_array)
print(random_array)
```

```python
ravelled_array = random_array.ravel()
ravelled_array[3]=4
print(ravelled_array)
print(random_array)
```
