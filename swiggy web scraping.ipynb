```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.swiggy.com/city/pune/best-restaurants"
headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
```

```python
response=requests.get(url, headers=headers)
if response.status_code==200:
    page_html=response.text
    print(f"page fetched sucessfully")
else:
    print(f"failed to fetch page")
    exit()
```

```python
page_html[:1000]
```

```python
page_soup=BeautifulSoup(page_html,"html.parser")
page_soup
```

```python
containers = page_soup.find_all("div",{'class':'styled__StyledRestaurantGridCard-sc-fcg6mi-0 lgOeYp'})
print(f'Number of containers found:{len(containers)}')

```

```python
if len(containers)>0:
    filename="sc-beySbM iKLEMo"
    f = open(filename,'w')

    
    headers_file='Name,Rating,Duration,location\n'
    f.write(headers_file)
    for container in containers:
        name_hotel=container.find_all("div",{"class":"sc-beySbM iKLEMo"})
        Name = name_hotel[0].text.strip() if name_hotel else 'N/A'


        rating_hotel=container.find_all("span",{"class":"sc-beySbM jAtOQO"})
        Rating = rating_hotel[0].text.strip().split("•")[0].strip() if rating_hotel else 'N/A'


        duration_hotel=container.find_all("div",{"class":"sw-restaurant-card-subtext-container"})
        Duration = duration_hotel[0].find("div", {"class": "sc-beySbM brneVe"}).text.strip().split("•")[1].strip() if duration_hotel else 'N/A'

        location_hotel=container.find_all("div",{"class":"sc-beySbM bRTXBF"})
        location = location_hotel[0].text.strip() if location_hotel else 'N/A'

        print(f"{Name},{Rating},{Duration}")
        f.write(f"{Name},{Rating},{Duration}\n")
    
    f.close()
```
