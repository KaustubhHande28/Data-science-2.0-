```python
help(KMeans)
```

```python
cluster_labels= model.fit_predict(scaled_X)
```

### Choosing the right K value 

```python
#creat a function that takes different values of k (minimum 2) and run KMeans model
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
ssd = []

for k in range (2,15):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,15),ssd,'*--')
plt.xlabel('K Value')
plt.ylabel('Sum od squared Distance')
plt.show()
```

```python
ssd
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```

```python
pd.Series(ssd).diff()
```

### Color Quantization using KMeans

```python
 import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_as_array = mpimg.imread('istockphoto-478656454-612x612.jpg')
```

```python
image_as_array
```

```python
plt.imshow(image_as_array)
```

```python
(h,w,c)=image_as_array.shape  #(h,w,c)color channels
```

```python
image_as_array_2d=image_as_array.reshape(h*w,c)
image_as_array_2d
```

```python
model= KMeans(n_clusters= 2)
labels = model.fit_predict(image_as_array_2d)
```

```python
#it is the 2 rgb cluster code
```

```python
model.cluster_centers_
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
quantized_image= np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantized_image
plt.imshow(quantized_image)
```
